```python
import numpy as np

def standardize(x):
    """
    Return a standardized copy of x: (x - mean) / std
    Does not mutate the input.
    """
    x = np.asarray(x, dtype=float)
    mean = x.mean()
    std = x.std()

    if std == 0:
        raise ValueError("Standard deviation is zero")

    return (x - mean) / std
```

```python
data = np.array([1, 2, 3, 4, 5])
z = standardize(data)

print("original:", data)
print("standardized:", z)
```
